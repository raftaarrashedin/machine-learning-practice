```python
# stemmer = nltk.SnowballStemmer("english")
def roman_urdu_stemmer(word):
    # List of common suffixes in Roman Urdu
    suffixes = ['i', 'e', 'on', 'ing', 's', 'es', 'ed', 'er', 'est', 'ai', 'ane', 'anein', 'iyat', 'iyon', 'iat', 'iyat', 'hon', 'hoon', 'iyon', 'ion', 'iyon', 'ne', 'ein', 'on', 'a', 'e', 'i', 'u', 'y', 'ae', 'ya', 'ao', 'yoon', 'oo', 'iyoon']

    for suffix in suffixes:
        if word.endswith(suffix):
            word = word[:-len(suffix)]
            break

    return word
```
