```python
#K-Means Clustering

#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
#allow charts to appear in the notebook
%matplotlib inline
```

```python
#importing the dataset
dataset = pd.read_csv('K_means_Algo.csv')
dataset
# df = pd.read_csv('kmeans.csv')
# df.head()
```

```python
X = dataset.iloc[:, :-1].values
print(X)
```

```python
print(dataset.columns)
```

```python
X = dataset[dataset.columns[:-1]]
X = X._convert(numeric=True)
#X = X.convert_objects(convert_numeric=True)
X.head()
```

```python
# eliminating null values
for i in X.columns :
  X[i] = X[i].fillna(int(X[i].mean()))

for i in X.columns :
  print(X[i].isnull().sum()) 


# converting 'Field_2' from float to int
```

```python
#using the elbow method to find the optimal number of clusters

# from sklearn.cluster import KMeans
wcss = []

# X = dataset[dataset.columns[:-1]]

for i in range(0,11):
  kmeans = KMeans(n_clusters=,init ='k-means++',max_iter=300,n_init=10,random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(0,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()



# these code is not working find solution, ohkieeee
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng :
  km = KMeans(n_clusters = k)
  km.fit(dataset[['cylinders','cubicinches','hp','time-to-60','year']])
  sse.append(km.inertia_)


plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.plot(k_rng,sse)
```
