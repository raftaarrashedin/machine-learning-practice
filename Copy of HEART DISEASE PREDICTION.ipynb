Importing the Dependencies

```python
#importing the dependencies
import numpy as np 
#this numpy libraries is used to numpy in array

import pandas as pd 
# pandas used  to create data frame 

from sklearn.model_selection import train_test_split #
#then we're importing this , we need to split our original data into training and test data

from sklearn.linear_model import LogisticRegression
#then we're importing logistic regression for binary values

from sklearn.metrics import accuracy_score 
# finally we're importing our accuracy score , so this accuracy score is evaluate that how well our model 
```

Data Collection and Processing

```python
# 1 loading the csv data to a oandas DATAFRAME
heart_data = pd.read_csv('/content/heart.csv')
```

```python
#print first five rows of the dataset 
heart_data.head()
```

```python
# the number of rows and columns 
heart_data.shape
```

```python
#last 5 rows from dataset
heart_data.tail()
```

```python
#getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statictical measures about the data
heart_data.describe()
```

```python
#checking the distribution of target variables
heart_data['target'].value_counts()
```

1 -----------> Defective Heart

0 -----------> Healthy Heart

--------------------------------------------------------
Splitting the Features and Target

```python
X = heart_data.drop(columns ='target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Sore

```python
# so we have imported that accuracy score from sklearn library,
# accuracy score is nothing but the model will be asked to predict the target
# and this predict value will compare with original target value, if model is predicted 
# about 99 values are correct out of 100, in that case the accuracy score is 99%.

# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```

Building a Predicted System

```python
# take input array
input_data = (55,1,0,132,353,0,1,132,1,1.2,1,1,3)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for only one instance 
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
  print('The Person is healthy and does not have a Heart Disease.')
else :
  print('The Person has a heart disease.')
```
