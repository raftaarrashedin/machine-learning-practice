```python
import numpy as np
```

```python
a = np.array([[10,120]])
b = np.array([[30,40]])
print(a)
print(b)
print()

(a+b).T
```

```python
#subtracting 2 vectors
(a - b)
(a - b).T
```

```python
#scaler multiplication
a = 2
u = np.array([[30,32]])
(a * u).T
```

```python
# dot product of 2 vectors
#for 2-D vectors, it is the equivalent to matrix multiplication

u = np.array([10,20])
v = np.array([38,39])

print(10 * 38 + 20 * 39)
np.dot(u,v)
```

```python
#matrix addition
a = np.array([[12,10],[4,5],[30,40],])
b = np.array([[2,8],[46,63],[50,10],])

print(a, end="= a \n\n")
print(b, end= "= b \n\n")

a + b
```

```python
#matix subtraction
a - b
```

```python
#Matrix scalar multiplication
a = 4
b = np.array([[2,5],[3,7],[8,6],])

a * b
```

```python
a = [[5,10],[15,20]]
b = [10,12]

np.dot(a,b)
```

```python
#Transpose a matrix
a = np.array([[5,10],[15,20],[25,30]])
print(a, end = "= a \n\n")
print(a.T, end = "=aT")
```

```python
#crearing an identity matrix

i1 = np.eye(2)
i2 = np.eye(3)
print(i1,end = "I1 \n\n")
print(i2,end = "I2 \n\n")
```

```python
#Inverse a matrix
a = np.array([
    [2,10,3],
    [29,44,65],
    [32,54,75],
])
print(a,end = " = a \n\n")
print(np.linalg.inv(a), end = " =  |a| ")
```
