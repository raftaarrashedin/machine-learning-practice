```python
# probability set 
from itertools import product as prod

A = set([x for x in range(1,7)])
B = set([x for x in range(1,7)])

p = list(prod(A,B))

print("A is set of all possible outcomes of a dice : ",A)
print("B is set of all possible outcomes of a dice : ",B)

print("\n Product of A and B is the all possible combinations of A and B thrown together : \n",p)
```

```python
# Throwing 2 dice

n_dice = 2
dice_faces = {1,2,3,4,5,6}

#Construct the event space i.e. set of ALL POSSIBLE events
event_space = set(prod(dice_faces, repeat = n_dice))
event_space
```

```python
for outcome in event_space :
  print(outcome, end = ' , ')

print()
print(len(event_space))
```

```python
# Sum is a multiple of 3

favorable_outcome = []
for outcome in event_space :
  x,y = outcome 
  if(x+y)%3 == 0:
    favorable_outcome.append(outcome)

favorable_outcome = set(favorable_outcome)

for f_outcome in favorable_outcome :
  print(f_outcome, end = ' , ')

print()
print(len(favorable_outcome))

prob = len(favorable_outcome) / len(event_space)
print("The probability of getting a sum which is a multiple of 3 is : ",prob)
```

```python
# Throwing 5 dice

n_dice = 5
dice_faces = {1,2,3,4,5,6}

#Construct the event space i.e., set of ALL POSSIBLE events
event_space = set(prod(dice_faces,repeat = n_dice))
event_space
```

```python
len(event_space)
```

```python
# Sum is a multiple of 5 but not a multiple of 3

favorable_outcome = []
for outcome in event_space :
  d1,d2,d3,d4,d5 = outcome
  if(d1+d2+d3+d4+d5) % 5 == 0 and (d1+d2+d3+d4+d5) % 3 != 0 :
    favorable_outcome.append(outcome)

favorable_outcome = set(favorable_outcome)
print("Length is : ",len(favorable_outcome))
prob = len(favorable_outcome)/len(event_space)
print("The probability of getting a sum, which is a multiple of 5 but not a mutiple of 3 is : ",prob)
print(favorable_outcome)
```

```python
# calculating conditional probability 
# let's find out the chances of a person getting lung disease due to smoking
pA = float(input("Prior probability of " "% people having lung disease : "))
pB = float(input("Probability that a patient smokes is " "% : "))
pBA = float(input("Probability of patients smokes even though they have lung disease is " "%"))
pAB = (pBA * pA) / pB
print("Probability that the patient will have lung disease if they smoke : ",pAB)
percentage = abs(pAB)
print("Hence, there is a {}% chance that a person who smoke will get a lung disease.".format(percentage))
```
