Use context manager(with)


*   It's a good idea to close a file after usage as it will free up the resources.
*   If we dont close it, garbage collector would close it.
* with keyword closes the file as soon as the usage is over



```python
# with
with open('/content/kamyabi.txt','w') as f :
  f.write('selmon bhai, tm kesi ho , kaha rehti ho kkch nh pt aur na hi sb mann krta h ye sb puchne ka, bsss mann krta h jesa chal raha h chalne dete h dukh is baat ka h, yaadein yhi chuti h tumhare h aur tm chali gyi inko chhor kr mere liye,khair jeee lenge ab strong ban chuke h')
```

```python
# try f.read() now

with open('/content/kamyabi.txt','r') as f :
  print(f.readline())
```

```python
# moving within a file -> 10 char then 10 char
with open('/content/kamyabi.txt','r') as f :
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
```

```python
# benifit -> to a load big file in the memory
big_file = ['hello world ' for i in range(1000)]

with open('kamyabi.txt','w') as f :
  f.writelines(big_file)
```

```python
with open('kamyabi.txt','r') as f :

  chunk_size = 10

  while len(f.read(chunk_size)) > 0 :
    print(f.read(chunk_size), end='###')
    f.read(chunk_size)
```

Serializing and Deserializing custom objects

```python
class person :

  def __init__(self,fname,lname,age,gender) :

    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender
```

```python
pearson = person('Raftaar','Rashedin',23,'Male')
```

```python
# as a string
import json

def show_object(person) :
  if isinstance(person,pearson) :
    return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

with open('example_1.json','w') as f :
  json.dump(person,f, default=show_object)
```

```python
# as a dict
import json

def show_object(pearson) :
  if isinstance(pearson,person) :
    return {'name':person.fname + ' ' + person.lname,'age' : person.age,'gender' : person.gender}


with open('example_1.json','w') as f :
  json.dump(person,f,default=show_object,indent=4)
```

```python
# deserialization
import json

with open('example_1.json','r') as f :
  d = json.load(f)

  print(d)
  print(type(d))
```

**Pickling**

Pickling is the process whereby a python object hierarchy is converted into a byte system, and unpickling is the inverse operation, whereby a byte stream ( from a binary file or bytes like object) is converted back into an object hierarchy.

```python
class Person :

  def __init__(self,name,age) :
    self.name = name
    self.age = age

  def display_info(self) :
    print('Hi my name is', self.name, 'and my age is ',self.age)
```

```python
p = Person('raftaar',23)
```

```python
# pickle dump
import pickle
with open('person.pkl','wb') as f :
  pickle.dump(p,f)
```

```python
# pickle load
import pickle

with open('person.pkl','rb') as f :
  p = pickle.load(f)

p.display_info()
```
