  List Vs Numpy
  

```python
from time import process_time
```

Time taken by a List

```python
python_list =  [i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end = process_time()

total_time_List = end - start_time
print(total_time_List)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5 

end_time = process_time()

total_time_numpy = end_time - start_time
print(total_time_numpy)
```

```python
if(total_time_List > total_time_numpy) :
  print('List consume more time')
  print('Numpy array is much Faster')
else :
  print('Numpy array consume more time')
  print('List is faster than Numpy')
```
