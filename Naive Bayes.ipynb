Here we will work on Titanic Dataset


```python
# Importing the Libraries 
import pandas as pd
df = pd.read_csv('titanic_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 'columns', inplace=True)
df.head()
```

```python
# df['Survived'] = df['Survived'].astype('Int64')
```

```python
target = df.Survived
input = df.drop('Survived', axis = 'columns')
```

```python
dummies = pd.get_dummies(input.Sex)
dummies.head()
```

```python
input.head()
```

```python
input = pd.concat([input,dummies], axis = 'columns')
input.head()
```

```python
input.drop('Sex', axis='columns', inplace = True)
input.head()
```

```python
input.columns[input.isna().any()]
```

```python
input.Age[:10]
```

```python
input.Age = input.Age.fillna(input.Age.mean())
# input.pclass = input.pclass.fillna(input.pclass.mean())
# input.fare = input.fare.fillna(input.fare.mean())
input.head()
```

```python
target.head()
```

```python
input.columns[input.isna().any()]
```

```python
import numpy as np

#No need to replace nan. You can pass to Int64 safely by doing:

# input['Pclass'] = np.floor(pd.to_numeric(input['Pclass'], errors='coerce')).astype('Int64')

input['Age'] = np.floor(pd.to_numeric(input['Age'], errors='coerce')).astype('Int64')

input['Fare'] = np.floor(pd.to_numeric(input['Fare'], errors='coerce')).astype('Int64')
```

```python
target.head()
```

```python
input.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train,y_test = train_test_split(input, target, test_size = 0.2)
```
